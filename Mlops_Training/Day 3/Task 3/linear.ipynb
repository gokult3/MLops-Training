**Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

**Load Dataset**

```python
df = pd.read_csv('/canada_per_capita_income.csv')
df.head()
```

**Plot Data**

```python
plt.scatter(df['year'], df['per capita income (US$)'])
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Canada Per Capita Income")
plt.show()
```

**Prepare data**

```python
X = df[['year']]
y = df['per capita income (US$)']
```

**Create and train linear regression model**

```python
model = LinearRegression()
model.fit(X, y)
```

**Predict income of year 2020**

```python
prediction = model.predict([[2020]])

print("Predicted per capita income in 2020:", prediction[0])
```

**Visualisation - Regression Line**

```python
plt.scatter(df['year'], df['per capita income (US$)'])
plt.plot(df['year'], model.predict(X))
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Regression Line")
plt.show()
```

**Check slope and intercept**

```python
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```
