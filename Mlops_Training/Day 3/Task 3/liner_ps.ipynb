```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# dataset
df = pd.DataFrame({
    'year': [1970,1980,1990,2000,2010,2016],
    'income': [3399,7000,12000,24000,35000,39000]
})

print(df)

# graph
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year, df.income, color='red', marker='+')
plt.show()

# input
X = df[['year']]

# output
y = df['income']

# model
model = LinearRegression()

# train
model.fit(X, y)

# predict 2020 income
prediction = model.predict([[2020]])

print("\nPredicted Canada Per Capita Income in 2020:")
print(prediction[0])

# coefficient
print("\nCoefficient:")
print(model.coef_)

# intercept
print("\nIntercept:")
print(model.intercept_)
```
